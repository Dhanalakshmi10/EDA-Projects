### 2) It has been observed that one/few outliers are present in the variable â€˜SOLD PRICEâ€™.
i) Draw a box plot for â€˜SOLD PRICEâ€™ with outliers & find the value of IQR, Upper limit & lower limit. 

```python
#box plot using seaborn
plt.figure(figsize=(8,5))
sns.boxplot(x="SOLD PRICE",data=df)
plt.show()
```

```python
#box plot using matplotlib
plt.figure(figsize=(10,8))
outlier_shape=dict(marker='+')
plt.boxplot(x="SOLD PRICE",data=df,vert=False, flierprops=outlier_shape)
plt.show()
```

```python
# IQR calculation
Q1 = df["SOLD PRICE"].quantile(0.25) 
Q3 = df["SOLD PRICE"].quantile(0.75)   
IQR = Q3 - Q1
Upper_Limit = Q3 + 1.5*IQR
Lower_Limit = Q1 - 1.5*IQR
print(f'Interquartile Range of Sold Price is {IQR} INR')
print(f'Sold price Upper Limit is {Upper_Limit} INR and Lower limit is {Lower_Limit} INR')
```

ii) Remove outliers & draw the box plot.

```python
outlier_df=df.copy() # creating a copy df as i do not want to make changes to my working df
Without_Outliers=outlier_df[(outlier_df['SOLD PRICE']>=Lower_Limit) & (outlier_df['SOLD PRICE']<=Upper_Limit)]
Without_Outliers
#we see that initially there were 130 entries, now there are only 122. 8 values have been removed.
```

```python
sns.boxplot(y='SOLD PRICE',data=Without_Outliers)
#no outliers are seen
```
